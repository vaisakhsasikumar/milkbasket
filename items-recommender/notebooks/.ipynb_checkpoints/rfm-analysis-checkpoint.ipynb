```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from lifetimes.utils import *
from lifetimes import BetaGeoFitter,GammaGammaFitter
from lifetimes.plotting import plot_probability_alive_matrix, plot_frequency_recency_matrix, plot_period_transactions, plot_cumulative_transactions,plot_incremental_transactions
from lifetimes.generate_data import beta_geometric_nbd_model
from lifetimes.plotting import plot_calibration_purchases_vs_holdout_purchases, plot_period_transactions,plot_history_alive
```

```python
orders = pd.read_csv('../data/hackathon_data.csv')
items = pd.read_csv('../data/hackathon_data.csv')
cust = pd.read_csv('../data/hackathon_data.csv')
orders.describe()
```

### Data Segregation from Single source

```python
print(orders.columns)
#cust.groupby('customer_id').size().value_counts()
cust = cust[['customer_id','society_id','route_id','city_id']]
items = items[['order_id','product_id','product_quantity','selling_price_per_unit']]
orders.head()
orders =  orders[['order_id','customer_id','order_date','product_addedtobasket_on','selling_price_per_unit']]

#orders.columns
```

```python
items.drop_duplicates('order_id',keep='first',inplace=True)
```

```python
transaction_data = pd.merge(orders,items,'inner','order_id')
transaction_data = transaction_data[['customer_id','order_date','selling_price_per_unit_y']]
# convert timestamp to date; only need the day
transaction_data['date'] = pd.to_datetime(transaction_data['order_date']).dt.date
transaction_data = transaction_data.drop('order_date',axis=1)
transaction_data.head()
```

#### From transaction data,we convert this into dataframe with the RFM variables. The Lifetimes package has a function for converting transaction data into an RFM DataFrame.


```python
summary = summary_data_from_transaction_data(transaction_data,'customer_id','date',monetary_value_col='selling_price_per_unit_y',)
summary.describe()
```

* frequency: # of days in which a customer made a repeat purchase
* T: customer's age in days
* recency: customer's age in days at time of most recent purchase
* monetary_value: mean of a customer's purchases, excluding the 1st purchase
#### frequency excludes the customer's first purchase because that is considered the day the customer is born. Afterwards, you can begin to question whether or not that customer is alive.

```python
bgf = BetaGeoFitter(penalizer_coef=14)
bgf.fit(summary['frequency'], summary['recency'], summary['T']);
plot_frequency_recency_matrix(bgf);
```

```python
plot_period_transactions(bgf).set_yscale('log');

```

```python
df = summary[summary['frequency']>0]
df['prob_alive'] = bgf.conditional_probability_alive(df['frequency'],df['recency'],df['T'])
sns.distplot(df['prob_alive']);
```

```python
df['churn'] = ['churned' if p < .1 else 'not churned' for p in df['prob_alive']]
sns.countplot(df['churn']);
```

```python
sns.distplot(df[df['churn']=='not churned']['prob_alive']).set_title('Probability alive, not churned');
```

```python
plot_incremental_transactions(bgf, transaction_data, 'date', 'customer_id', 450, 365);
```

```python
summary_cal_holdout = calibration_and_holdout_data(transaction_data, 'customer_id', 'date',calibration_period_end='2019-06-30', observation_period_end='2019-07-30' )
plot_calibration_purchases_vs_holdout_purchases(bgf, summary_cal_holdout);
```

```python
df['churn'][(df['prob_alive']>=.1) & (df['prob_alive']<.2)] = "high risk"
df['churn'].value_counts()
```
